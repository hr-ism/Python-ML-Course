```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X- Label')
plt.ylabel('Y- Label')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.35,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')


axes2.plot(y,x,'r')
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows = 1,ncols =2)
axes[0].plot(x,y,'g')
axes[1].plot(y,x,'d')
```

```python
fig,axes = plt.subplots(nrows = 2,ncols =1,figsize =(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = 'X sqaured')
ax.plot(x,x**3,label = 'X cubed')
ax.legend()
```
